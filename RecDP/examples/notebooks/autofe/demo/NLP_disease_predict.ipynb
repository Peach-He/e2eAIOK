```python
from pyrecdp.datasets import CESM_breast_cancer
data = CESM_breast_cancer().to_pandas()
```

```python
CESM_diagnose = data['manual_annotations'][data['manual_annotations']['Type'] =='CESM']
```

```python
train_df = CESM_diagnose.merge(data['medical_report'], on = ['Patient_ID', 'Side', 'Type'], how='left')
train_df
```

```python
train_df = train_df.groupby(['Patient_ID', 'Side']).agg({'Pathology Classification/ Follow up':lambda x: list(x), })
train_df
```
